```python
import pandas as pd;
import numpy as np;
import seaborn as sb;
import matplotlib.pyplot as plt;
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding = 'unicode_escape')
df
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.rename(columns={'Shaadi':'Marital_Status'},inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df[['Age','Amount','Orders']].describe()
```

```python
# Exploratory Data Analysis
# Gender
```

```python
df.columns
```

```python
sb.countplot(x= 'Gender', data = df)
```

```python
ax = sb.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
Sales_gender = df.groupby(['Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount', ascending=False)
Sales_gender
```

```python
gen = sb.barplot(x='Gender', y='Amount',data = Sales_gender)
for bars in gen.containers:
    gen.bar_label(bars)
```

```python
# From above graph we come to know that most of the buyers are females and the purchasing power of females is greater than men.     
```

```python
# Age
```

```python
ax = sb.countplot(x='Age Group', data = df, hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_age = df.groupby(['Age Group'], as_index =False) ['Amount'].sum().sort_values(by ='Amount',ascending=False)
```

```python
sb.barplot(x='Age Group',y='Amount', data=Sales_age)
```

```python
#Most of the buyers are from the age group 26-35 females
```

```python
#State
```

```python
Sales_state = df.groupby(['State'], as_index =False) ['Orders'].sum().sort_values(by ='Orders',ascending=False).head(15)
sb.set(rc={'figure.figsize':(25,5)})
sb.barplot(data = Sales_state, x = 'State', y = 'Orders')
```

```python
Sales_state = df.groupby(['State'], as_index =False) ['Amount'].sum().sort_values(by ='Amount',ascending=False).head(15)
sb.set(rc={'figure.figsize':(25,5)})
sb.barplot(data = Sales_state, x = 'State', y = 'Amount')
```

```python
#The above two graphs shows us that most orders and total sales/amount cones from the states of Uttar Pradesh, Maharashtra and Karnataka 
```

```python
ax = sb.countplot(x = 'Marital_Status', data = df )
for bars in ax.containers:
    ax.bar_label(bars)
sb.set(rc={'figure.figsize':(7,7)})
```

```python
Sales_Marital_status = df.groupby(['Marital_Status', 'Gender'], as_index =False) ['Amount'].sum().sort_values(by ='Amount',ascending=False)
sb.barplot(data = Sales_Marital_status, x = 'Marital_Status', y = 'Amount', hue='Gender')
```

```python
#From the above data we come to know that most of the buyers are married(Women) and have high purchasing power
```

```python
#Occupation
```

```python
sb.set(rc={'figure.figsize':(20,5)})
ax = sb.countplot( data = df , x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Occupation = df.groupby(['Occupation'], as_index =False) ['Amount'].sum().sort_values(by ='Amount',ascending=False).head(15)
sb.set(rc={'figure.figsize':(25,5)})
ax = sb.barplot(data = Sales_Occupation, x = 'Occupation', y = 'Amount')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#From the above graphs we can see that most of the buyers are from the IT, Healthcare and Aviation sector
```

```python
#Product Category
```

```python
sb.countplot(data = df , x = 'Product_Category')
```

```python
Sales_Product = df.groupby(['Product_Category'], as_index =False) ['Amount'].sum().sort_values(by ='Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(20,5)})
ax = sb.barplot(data = Sales_Product, x = 'Product_Category', y = 'Amount')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#From the above graph we can see that food, clothing and electronic products are sold the most
```

```python
Sales_Product_ID = df.groupby(['Product_ID'], as_index =False) ['Amount'].sum().sort_values(by ='Amount',ascending=False).head(10)
sb.set(rc={'figure.figsize':(25,5)})
ax = sb.barplot(data = Sales_Product_ID, x = 'Product_ID', y = 'Amount')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#The top 10 most selling products are mentioned above 
```

```python
#Conclusion
#Married women from the age group 26-35 years who work in the field of IT, Healthcare and Aviation are more likely to buy products from food, clothing and electronics
```
